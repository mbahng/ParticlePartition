# Read the eventHits.txt file and count the number of objects

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Wedge_Object Reader

```python
def read_file(file_name):
    wedge_objects = []
    try:
        with open(file_name, 'r') as file:
            lines = file.readlines()
        for line in lines:
            wedge_objects.append(line.strip())
        return wedge_objects
    except FileNotFoundError:
        print(f"File '{file_name}' not found.")
        return None

# Usage example
file_name = "wedge_objects_128.txt"
wedge_objects = read_file(file_name)
# if wedge_objects:
#     for i, wedge_object in enumerate(wedge_objects, start=1):
#         print(f"wedge_object_{i}: {wedge_object}")
```

```python
# # Specify the output file path
# output_file = "Wedged_Spacepoints_12.txt"

# # Iterate over the wedge object spacepoints and write them to the file
# with open(output_file, "w") as f:
#     for key, value in wedge_object_spacepoints.items():
#         line = f"{key}: {value}\n"
#         f.write(line)
```

```python
import numpy as np

# Create a dictionary to store spacepoints for each wedge object
wedge_object_spacepoints = {}

# Iterate over each event
for event_number in range(1, 51):
    # Initialize a dictionary to store spacepoints for each wedge object in the current event
    event_spacepoints = {}

    # Iterate over each wedge object
    for i in range(len(wedge_objects)):
        # Initialize a list to store extracted spacepoints for the current wedge object
        spacepoints = []

        # Extract phi_start and phi_end values for the current wedge object
        phi_start_values_i = phi_start_values[f'phi_start_{i+1}']
        phi_end_values_i = phi_end_values[f'phi_end_{i+1}']

        event = globals()[f'event_{event_number}']  # Get the event data for the current event number

        # Iterate over each spacepoint in the current event
        for spacepoint in event:
            layer = spacepoint[0]  # Extract layer number
            phi = spacepoint[2]  # Extract Phi angle

            # Adjust phi_start and phi_end values for angles greater than 2*pi radians
            adjusted_phi_start_values_i = [phi_start % (2.5 * np.pi) for phi_start in phi_start_values_i]
            adjusted_phi_end_values_i = [phi_end % (2.5 * np.pi) for phi_end in phi_end_values_i]

            # Check if the Phi angle falls within the corresponding range for the layer
            if any(
                (start <= phi <= end) or (start <= phi + 2 * np.pi <= end) or (start <= phi - 2 * np.pi <= end)
                for start, end in zip(adjusted_phi_start_values_i, adjusted_phi_end_values_i)
            ):
                spacepoints.append(spacepoint)  # Add the spacepoint to the spacepoints list

        # Store the spacepoints in the current wedge object dictionary
        event_spacepoints[f'wedge_object_{i+1}'] = spacepoints

    # Store the event spacepoints in the main dictionary
    wedge_object_spacepoints[f'event_{event_number}'] = event_spacepoints
```

```python
output_file = "wedgeData_v3_128.txt"

# Write all spacepoints to the file
with open(output_file, "w") as f:
    for event_key, event_value in wedge_object_spacepoints.items():
        for wedge_key, wedge_value in event_value.items():
            line = f"{', '.join([str(x).replace(' ', '') for x in wedge_value]).replace(', ', ',')}\n"
            if line.strip():  # Check if the line is not empty
                f.write(line)
        f.write("\n")

# Read the file and remove empty lines
with open(output_file, "r+") as f:
    lines = f.readlines()
    f.seek(0)
    f.writelines(line for line in lines if line.strip())
    f.truncate()
```
